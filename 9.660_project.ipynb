#Data Analysis: Bootstrapping

Steps:
 

1.   Take specific width and depth
2.   Sample responses (bootstrapping) to likely and severity questions
3.   Apply hypothesized function
4.   Plot distribution of risk
5.   Compare to actual distribution of risk



```python
num_samples = 5000
bootstrap_samples = 2
```

```python
def which_function(function, likely, severity):
  """
  calls function(likely, severity)
  """
  return function(likely, severity)

def just_likely(likely, severity):
  return likely

def just_severity(likely, severity):
  return severity

def multiply(likely, severity):
  """
  elementwise multiplication of likely and severity
  """
  return list(np.multiply(np.array(likely), np.array(severity)))

def add(likely, severity):
  """
  elementwise addition of likely and severity
  """
  return list(np.add(np.array(likely), np.array(severity)))

def elem_min(likely, severity):
  """
  elementwise minimum of likely and severity
  """
  return list(np.minimum(np.array(likely), np.array(severity)))

def elem_max(likely, severity):
  """
  elementwise maximum of likely and severity
  """
  return list(np.maximum(np.array(likely), np.array(severity)))

def geometric_mean(likely, severity):
  """
  square root of elementwise multiplication of likely and severity
  """
  return list(np.sqrt(np.multiply(np.array(likely), np.array(severity))))

def avg(likely, severity):
  """
  elementwise average of likely and severity
  """
  return list(np.add(np.array(likely), np.array(severity))/2)

def multiply_scaled(likely, severity,):
  """
  elementwise multiplication scaled by dividing by 7 of likely and severity
  """
  return list(np.multiply(np.array(likely), np.array(severity))/7)

def just_one_based_on_val(likely, severity):
  """
  for small differences in likely and severity rating, uses averages
  for small values of likely or severity when other is very different, uses itself
  for large values of likely or severity when other is different, uses opposite
  """
  my_list = []
  avg_likely = np.mean(likely)
  avg_severity = np.mean(severity)
  #print("avg likely = ", avg_likely, ", avg severity = ", avg_severity)
  if (avg_severity>5 and avg_likely>5) or (avg_severity <2 and avg_likely <2):
    #print("used average: both high or low")
    return avg(likely, severity)
  elif avg_severity > 5:
    #print("used likely: severity > 5")
    return just_likely(likely, severity)
  elif avg_likely > 5:
    #print("used severity: likely > 5")
    return just_severity(likely,severity)
  elif avg_severity < 2:
    #print("used severity: severity < 2")
    return just_severity(likely, severity)
  elif avg_likely <2:
    #print("used likely: likely <2")
    return just_likely(likely, severity)
  else:
    #print("used average")
    return avg(likely, severity)

def geometric_mean_and_min(likely, severity):
  avg_likely = np.mean(likely)
  avg_severity = np.mean(severity)
  if abs(avg_likely - avg_severity) <4:
    return geometric_mean(likely, severity)
  else:
    return elem_min(likely, severity)

def log_multiply(likely, severity):
  return list(np.log(np.multiply(np.array(likely), np.array(severity))))

def normalized_multiply(likely, severity):
  new_likely = [(i-1)/6 for i in likely]
  new_severity = [(j-1)/6 for j in severity]
  multiplied = list(np.multiply(np.array(new_likely), np.array(new_severity)))
  return [k*6+1 for k in multiplied]

def normalized_likelihood_multiply(likely, severity):
  new_likely = [i/7 for i in likely]
  multiplied = list(np.multiply(np.array(new_likely), np.array(severity)))
  return multiplied

possible_functions = [just_likely, just_severity, multiply, add, elem_min, elem_max, 
                      geometric_mean, avg, multiply_scaled, just_one_based_on_val, 
                      log_multiply, normalized_multiply, geometric_mean_and_min, normalized_likelihood_multiply]
```

```python
def create_risk_model(scenario, function):
  """
  scenario = type string depth_width       ex: shallow_thin
  function = type function                 ex: multiply
  """
  likely = my_dict[scenario + "_likely"]
  severity = my_dict[scenario + "_severity"]

  num_responses = len(likely)

  likely_sampling = []
  severity_sampling = []

  for _ in range(num_samples):
    total_sum_likely = 0
    total_sum_severity = 0
    for i in range(bootstrap_samples):
      total_sum_likely+= likely[random.randint(0, num_responses - 1)]
      total_sum_severity+= severity[random.randint(0, num_responses - 1)]
    likely_sampling.append(total_sum_likely/bootstrap_samples)
    severity_sampling.append(total_sum_severity/bootstrap_samples)

  risk_model_results = which_function(function, likely_sampling, severity_sampling)
  return risk_model_results

```

```python
def plotting(model, scenario_name):
  actual= my_dict[scenario_name+"_risk"]
  all_vals = model + actual
  min_val = int(min(all_vals))
  max_val = math.ceil(max(all_vals))
  f,a = plt.subplots(1,2)
  a[0].hist(model, bins = [0.5 + x for x in range(min_val-1, max_val+1)])
  a[0].set_title("Predicted_" + scenario_name + "_risk")
  a[0].set_xticks(range(min_val-1,max_val+1))

  a[1].hist(actual, bins = [0.5 + x for x in range(min_val-1, max_val+1)])
  a[1].set_title("Actual_" + scenario_name + "_risk")
  a[1].set_xticks(range(min_val-1,max_val+1))
  plt.tight_layout()
```
